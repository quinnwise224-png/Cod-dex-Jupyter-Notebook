```python
import plotly.express as px
```

```python
fig = px.histogram(df, x="subscribers")
fig.show()
```

```python
#Subscribers counts are distributed through the histrogram.
```

```python
fig = px.pie(df, names="category", values="subscribers")
fig.show()

# Total subscribers are divided by category is through the pie chart.
```

```python
fig = px.box(df, y="subscribers")
fig.show()

# This box plot shows the spread of subscriber counts.
```

```python
fig = px.histogram(df, x="video_count", title="Distribution of Video Count")
fig.show()
```
